```python
# open the opening_moves.txt file and read each line
entries = []
with open('opening_moves.txt', 'r') as file:
    for line in file:
        entries.append(line.strip())

print(len(entries))
```

```python
# count the occurrence of each unique opening sequence
count = {}
for entry in entries:
    if entry in count:
        count[entry] += 1
    else:
        count[entry] = 1
```

```python
# print the number of occurrences of each unique opening sequence (in descending order)

plot = []
for i, (key, value) in enumerate(sorted(count.items(), key=lambda item: item[1], reverse=True)):
    print(i, value)
    plot.append(value)
```

```python
def get_total_for_range(start, end):
    total = 0
    for i in range(start, end):
        total += plot[i]
    return total

def get_total():
    return sum(plot)

def get_percentage_for_range(start, end):
    return get_total_for_range(start, end) / get_total() * 100

print(get_percentage_for_range(0, 256))
```

```python
# output the keys of count in descending order of value to a file
with open('sorted_opening_moves.txt', 'w') as file:
    for key, value in sorted(count.items(), key=lambda item: item[1], reverse=True):
        file.write(key + '\n')
```
